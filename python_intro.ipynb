Egy nagyon alapvető adatstruktúra a lista. A változóinkat sokszor nagyon hasznos listába rendezni.

```python
# python lista: dolgok egymas utan, zarojelek kozott
a = [0, 1, 2]     # Ha eddig más lett volna az a, most felülírjuk
print(a)

# python ciklus: be kell ugrasztani a sorokat, hogy lassuk, hol tartunk.
# Javasolt tabulatort hasznalni:

for j in [0, 1, 2]:
    print("Hello world", j)

    
# listaban barmi lehet
a = ["Hello", 3, 4.5]
for j in a:
    print("az a lista elemei:", j)
    
print("az a lista: ", a)

# lista elemei:
print(a[2])

```

Ciklusokhoz hasznos, ha egészek egy sorozatát is le tudjuk gyártani. Erre szolgál a "range". 

Nagyon hasznos a listák "szeletelése".


```python
# egeszek listajahoz jol jon a "range":
print(range(10))
for j in range(10):
    print(j)

# A range-et listava alakithatjuk a kicsomagolo * muvelettel
b = [*range(10)]

# lista szeletelheto - from:until:by
c = b[0:2]
d = b[:3]
print(b[1::2])
print("Visszafele: ", b[::-1])

for j in b[1::2]:
    print(j)

```

```python
import numpy as np

N = 115  # ennyi tagot veszunk a vegtelen helyett

pi_kozelites = []
sor = 0
for j in range(1, N+1):
    sor = sor + 1/j**2
    pi_kozelites.append(np.sqrt(6*sor))

for j in range(len(pi_kozelites)):
    print(pi_kozelites[j])
    
# BBP
sor = 0 
pi_BBP = []
for k in range(N):
    sor += 1/16**k * (4/(8*k+1) - 2/(8*k+4) -1/(8*k+5) - 1/(8*k+6))
    pi_BBP.append(sor)
```

```python
T = 100
kiserletek = []
for k in range(100):
    penz = 0
    p = 0.6
    vagyon = []
    for t in range(T):
        vagyon.append(penz)
        r = np.random.rand()
        penz -= 1
        if r<p: penz += 2
    kiserletek.append(vagyon)
    plt.plot(range(T), vagyon, alpha=0.1)
    
vart_vagyon = [2*(p-0.5)*j for j in range(T)]
        
plt.plot(range(T), vart_vagyon, "k")
plt.xlabel("Jatekok szama")
plt.ylabel("Vagyon") 

plt.show()
```

```python
# A numpy tomb (array) nagyon fontos adattipus. 
# Ez szamok tablazata. Tombon elemenkent lehet muveleteket vegezni.
# Tombot meg lehet szorozni szammal, hozza lehet adni szamot. 

a = np.array(range(10))
b = a+2
print("a = ", a)
print("b = a + 1 = ", b)
print(a + b[::-1])
print(a*b)
print(a**b)
print(np.cos(a)+b)
```

```python
# A fenti ermedobalos pelda is egyszerubb array hasznalataval:
T = 100  # hany jatek
k = 200  # hany kiserlet
p = 0.6

r = np.random.rand(k, T)
vagyon = np.cumsum(r<p, axis=1) - np.cumsum(p<r, axis=1)

#plt.matshow(vagyon)
#plt.colorbar()
#plt.show()

plt.plot(vagyon.T, alpha=0.1)
tt = np.arange(T)
plt.plot(2*(p-0.5)*tt, "k")
plt.plot(2*(p-0.5)*tt + np.sqrt(tt*p*(1.-p)), "k")
plt.plot(2*(p-0.5)*tt - np.sqrt(tt*p*(1.-p)), "k")
plt.xlim(0, T)

plt.show()


#vart_vagyon = [2*(p-0.5)*j for j in range(T)]
        
#plt.plot(range(T), vart_vagyon, "k")

#plt.show()
```
