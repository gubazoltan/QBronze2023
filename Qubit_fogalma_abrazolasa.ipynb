# Qubit (kétállapotú kvantumrendszer): 
- a klasszikus logikai bit kvantumos megfelelője, a kvantuminformáció alapegysége
- viselkedését a kvantummechanika törvényei írják le (lásd kvantumos időfejlődés, kvantumos mérés)
- (tiszta) állapota a $|0\rangle$ és $|1\rangle$ kvantumállapotok kombinációja:

&emsp;&emsp;&emsp;&emsp;&emsp; $|\psi\rangle=a|0\rangle+b|1\rangle=a\begin{pmatrix} 1 \\ 0 \end{pmatrix}+b\begin{pmatrix} 0 \\ 1 \end{pmatrix}=\begin{pmatrix} a \\ b \end{pmatrix}$ 

&emsp;&emsp;&emsp; ahol $a$ és $b$ az úgynevezett **amplitúdók** 

- (Megjegyzés: általában $a$ és $b$ komplex számok, de most csak azt fogjuk vizsgálni, amikor $a$ és $b$ valós)

- ha meg akarjuk ismerni a $|\psi\rangle$ kvantumállapotot, akkor meg kell mérnünk    

    - de **méréskor** csak vagy '0'-t vagy '1'-et kaphatunk, úgy, hogy
    
    &emsp;&emsp;&emsp;&emsp;&emsp; $P(0)=|a|^2$ &emsp; és &emsp; $P(1)=|b|^2$
    &emsp;&emsp;&emsp;&emsp;&emsp; 
    <br>
    
    - ugyanakkor a kettő közül valamelyiket biztosan kapjuk $\Rightarrow$ $P(0)+P(1)=1$
    
    &emsp;&emsp;&emsp;&emsp;&emsp; $\Rightarrow$ $|a|^2+|b|^2=1$ 
    <br>
    
    - azt mondjuk: a kvantumállapotoknak **normált**-aknak (vagy **egységnyi hosszúságúaknak**) kell lenniük

#### Kérdés: Az alábbi vektorok közül melyik tekinthető valódi kvantumállapotnak?

&emsp;&emsp;&emsp;&emsp;&emsp; $\begin{pmatrix} \frac{2}{\sqrt{3}} \\ \frac{1}{\sqrt{3}} \end{pmatrix}$
&emsp;&emsp;&emsp; vagy 
&emsp;&emsp;&emsp; $\begin{pmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{pmatrix}$ 

<br>

- Az első vektor esetében a két valószínűség $P(0)=\frac{4}{3}$ és $P(1)=\frac{1}{3}$, amiből $P(0)+P(1)=\frac{5}{3}$, tehát ez **nem valódi** kvantumállapot

- A második vektor esetében a két valószínűség $P(0)=\frac{1}{2}$ és $P(1)=\frac{1}{2}$, amiből $P(0)+P(1)=1$, tehát ez a vektor **valódi** kvantumállapot 
<br>

### Fontos még:
- $a$ és/vagy $b$ lehet negatív, de $|\psi\rangle$ és $-|\psi\rangle$ között nem teszünk különbséget! 
<br>

    - Például: $\begin{pmatrix} a \\ -b \end{pmatrix}$ és $\begin{pmatrix} -a \\ b \end{pmatrix}$ ugyanazt az állapotot jelölik

# Qubit kvantumállapotának szemléltetése

### Bloch-gömb valós főkörén

- A kvantumbitek (tiszta) állapotait általános esetben egy egységnyi sugarú gömbön lehet ábrázolni
- A valós amplitúdójú állapotok ennek az ún. Bloch-gömbnek az egyik főkörén helyezkednek el
- A Bloch-gömb északi pólusa a $|0\rangle$, a déli pólusa az $|1\rangle$ kvantumállapotnak felel meg, ezek rajta vannak a valós amplitúdójú állapotoknak megfelelő főkörön

![](https://github.com/gubazoltan/QBronze2023/blob/main/%C3%A1br%C3%A1k/Bloch_sphere_circle.png?raw=true)

- A valós $a,b$ amplitúdójú kvantumállapotokra $\varphi=0$, őket úgy paraméterezzük, hogy 

&emsp;&emsp;&emsp;&emsp;&emsp; $|\psi\rangle=a|0\rangle+b|1\rangle=\cos{\frac{\theta}{2}}|0\rangle+\sin{\frac{\theta}{2}}|1\rangle$ &emsp; ahol &emsp; $\theta\in\left[0,2\pi\right)$
&emsp;&emsp; azaz: 
&emsp;&emsp; $a=\cos{\frac{\theta}{2}}$, &emsp; $b=\sin{\frac{\theta}{2}}$

- Ezeket a Bloch-gömbön az alábbi koordinátákkal ábrázolhatjuk:

&emsp;&emsp;&emsp;&emsp;&emsp; $x_{\psi}=\sin{\theta}=2\sin{\frac{\theta}{2}}\cos{\frac{\theta}{2}}=2ab$,

&emsp;&emsp;&emsp;&emsp;&emsp; $y_{\psi}=0$,

&emsp;&emsp;&emsp;&emsp;&emsp; $z_{\psi}=\cos{\theta}=\cos^2{\frac{\theta}{2}}-\sin^2{\frac{\theta}{2}}=a^2-b^2$.

- Megadva ezeket a koordinátákat, a qiskitben ábrázolhatjuk a kvantumállapotokat a Bloch-gömbön is.

### Szemléltessünk qiskittel néhány kvantumállapotot a Bloch-gömbön!

Ehhez az alábbi függvényekre lesz szükségünk:

```python
from qiskit.visualization import plot_bloch_vector
from math import pi, sqrt, cos, sin
```

#### Ábrázoljuk az $\begin{pmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{pmatrix}=\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)$ állapotot! 

```python
# Nevezzük el "a"-nak és "b"-nek a kvantumállapot-vektor elemeit
a=1/sqrt(2)                                           
b=1/sqrt(2)

# Majd határozzuk meg a fenti képletekkel a kvantumállapot Bloch-koordinátáit.
x=2*a*b
y=0
z=a*a-b*b

plot_bloch_vector([x,y,z])
```

Tehát az $\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)$ állapot a Bloch-gömbön az $x$ tengely pozitív irányába mutat. 

Az $\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)$ állapotot ezért $|+\rangle$-nak szokás jelölni.

#### Most ábrázoljuk az $\begin{pmatrix} \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} \end{pmatrix}=\frac{1}{\sqrt{2}}(|0\rangle-|1\rangle)$ állapotot !

```python
# Nevezzük el "a"-nak és "b"-nek a kvantumállapot-vektor elemeit
a=1/sqrt(2)
b=-1/sqrt(2)

# Majd határozzuk meg a fenti képletekkel a kvantumállapot Bloch-koordinátáit.
x=2*a*b
y=0
z=a*a-b*b

plot_bloch_vector([x,y,z])
```

Tehát a $\frac{1}{\sqrt{2}}(|0\rangle-|1\rangle)$ állapot a Bloch-gömbön az $x$-tengely negatív irányába mutat.

Az $\frac{1}{\sqrt{2}}(|0\rangle-|1\rangle)$ állapotot ezért $|-\rangle$-nak szokás jelölni.

#### Feladat: Ábrázoljunk egy tetszőleges (de 1 hosszúságú!) vektort a Bloch-gömbön.

```python
#Ahhoz, hogy az állapot normált legyen, adjuk meg az amplitúdók négyzeteit úgy, 
#hogy összegük 1 legyen
a2=1/3
b2=1-a2

a=sqrt(a2)
b=sqrt(b2)

x=2*a*b
y=0
z=a*a-b*b

plot_bloch_vector([x,y,z])
```
