<h3> 1. Reprezentációk </h3>

Két-qubit állapotát felírhatjuk a következő bázisállapotok segítségével: $ \ket{00},\ket{01},\ket{10},\ket{11} $, ahol az első szám az első qubit, a második szám pedig a második qubit állapotát jelöli. A két kvantumbit egy tetszőleges állapota felírható a bázisállapotok lineáris kombinációjaként,

$$ \ket{\psi}=\alpha\ket{00}+\beta\ket{01}+\gamma\ket{10}+\delta\ket{11}. $$

A kvantumrendszer állapotát az absztrakt ket-vektorok helyett tradicionális vektorokkal is ábrázolhatjuk. Ilyenkor a bázisvektorok a következő alakban írhatóak fel:

$$ 
    \ket{00} = \myvector{1 \\ 0 \\ 0 \\ 0},
    ~~~~~~
    \ket{01} = \myvector{0 \\ 1 \\ 0 \\ 0},
    ~~~~~~
    \ket{10} = \myvector{0 \\ 0 \\ 1 \\ 0},
    ~~~~~~
    \ket{11} = \myvector{0 \\ 0 \\ 0 \\ 1}.
$$

A már korábban bevezetett általános $ \ket{\psi} $ vektor pedig a következő alakban áll elő,

$$ \ket{\psi}=\myvector{\alpha \\ \beta \\ \gamma \\ \delta}. $$

Egy kvantumbit állapotát egy 2 dimenziós, míg két kvantumbit állapotát egy 4 dimenziós vektorral tudjuk reprezentálni. Két független kvantumbit állapotát leíró 4 dimenziós vektort felírhatjuk a tenzorszorzás ($\otimes$) művelet segítségével. Két tisztán 0, vagy 1 állapotban lévő qubit tenzorszorzatával megkaphatjuk a bázisvektorokat.

$$ \ket{a} \otimes \ket{b} = \ket{ab}, \quad \text{ahol} \quad a,b \in \{0,1\} $$

Két tetszőleges 2 dimenziós vektor tenzorszorzatát a következő módon definiálljuk,

$$
  \myvector{a_1 \\ a_2} \otimes \myvector{b_1 \\ b_2}=
  \myvector{a_1b_1 \\ a_1b_2 \\ a_2b_1 \\ a_2b_2}.
$$

<h3> 2. Két-qubites operátorok </h3>

Az álábbi áramkör 2 qubitet és minden qubiten egy-egy Hadamard kaput tartalmaz.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

qc.draw(output='mpl',reverse_bits=True)
```

A Hadamard kapu egy egy-qubites kapu, aminek a hatását a bázisállapotokon a következő alakban írhatjuk fel:

$$ H \ket{0} = \dfrac{1}{\sqrt{2}}(\ket{0}+\ket{1}), \qquad H \ket{1} = \dfrac{1}{\sqrt{2}}(\ket{0}-\ket{1}). $$

Két Hadamard kapura gondulhatunk úgy is, mint egy két-qubites kapura. Ezt a kaput a két Hadamard kapu tenzorszorzataként írhatjuk fel,

$$ H \otimes H. $$

A $ H \otimes H $ kapu hatása megegyezik azzal, hogy a qubitekre egyenként alkalmazunk egy-egy Hadamard kaput. így a $ H \otimes H $ kapu a következő módon hat a bázisállapotokon:

$$ (H \otimes H)(\ket{a}\otimes \ket{b})=H\ket{a}\otimes H\ket{b}, \quad \text{ahol} \quad a,b \in \{0,1\}. $$

<h3> 3. Összefonódás </h3>

Két kvantumbit között felléphet az úgynevezett összefonódás jelensége. A két qubit állapotát akkor nevezzük összefonódottnak, ha az állapotuk nem írható fel két egy-qubites állapot tenzorszorzataként,

$$ \nexists \ket{\psi_1},\ket{\psi_2}, \quad \text{hogy} \quad \ket{\psi}=\ket{\psi_1}\otimes \ket{\psi_2}. $$

Például ilyen két-qubites állapotok az úgynevezett Bell állapotok is:

<ul>
    <li>$ \ket{\Phi^+} = \dfrac{1}{\sqrt{2}}(\ket{00}+\ket{11}) $, </li>
    <li>$ \ket{\Phi^-} = \dfrac{1}{\sqrt{2}}(\ket{00}-\ket{11}) $, </li>
    <li>$ \ket{\Psi^+} = \dfrac{1}{\sqrt{2}}(\ket{01}+\ket{10}) $, </li>
    <li>$ \ket{\Psi^-} = \dfrac{1}{\sqrt{2}}(\ket{01}-\ket{10}) $. </li>
</ul>

Az eredetileg nem összefonódott qubitek összefonását csak olyan két-qubites kapukkal lehet elérni, amelyek együtt hatnak a két kvantumbiten (nem írhatóak fel egy-qubites kapuk tenzorszorzataként). Ilyen a kapu a CNOT kapu is. Ez a kapu a következő módon hat a bázisállapotokon:

<ul>
    <li>$ CNOT \ket{00} = \ket{00} $, </li>
    <li>$ CNOT \ket{01} = \ket{01} $, </li>
    <li>$ CNOT \ket{10} = \ket{11} $, </li>
    <li>$ CNOT \ket{11} = \ket{10} $. </li>
</ul>

A CNOT kapu bázisállapotokon való hatását a qiskit segítségével is ellenőrízhetjük.

```python
pairs = ['00','01','10','11'] #A bázisállapotok

for pair in pairs:
    from qiskit import QuantumCircuit, execute, Aer
    qc = QuantumCircuit(2,2)
    # A bázisállapot inicializálása
    if pair[1] == '1':
        qc.x(0)
    if pair[0] =='1':
        qc.x(1)
    # CNOT kapu a két qubit között
    qc.cx(1,0)
    # A qubitek megmérése
    qc.measure(0,0)
    qc.measure(1,1)
    display(qc.draw(output='mpl',reverse_bits=True))
    # Az áramkör lefuttatása 1024-szer.
    job = execute(qc,Aer.get_backend('qasm_simulator'),shots=1024)
    counts = job.result().get_counts(qc)
    print(pair,"--CNOT->",counts)
```
